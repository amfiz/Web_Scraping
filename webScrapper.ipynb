**Question:1**

```python
import nltk
import re
from pprint import pprint

# Test Strings
test_strings = [
    "apple Banana",
    "A apple",
    "pat",
    "123.45",
    "xyz",
    "word123",
]

# Regular Expressions
regex_patterns = [
    r"[a-zA-Z]+",          # Matches one or more alphabetical characters (case insensitive)
    r"[A-Z][a-z]*",        # Matches an uppercase letter followed by zero or more lowercase letters
    r"p[aeiou]{,2}t",      # Matches 'p', followed by up to 2 vowels, and then 't'
    r"\d+(\.\d+)?",        # Matches digits with an optional decimal part
    r"([^aeiou][aeiou][^aeiou])*",  # Matches sequences of three characters where the middle one is a vowel
    r"\w+|[^\w\s]+",       # Matches words or sequences of non-alphanumeric characters
]

# Test and show results
for pattern in regex_patterns:
    print(f"Pattern: {pattern}")
    pprint(nltk.re_show(pattern, ' '.join(test_strings)))
    print("\n" + "="*50 + "\n")
```

```python
import re

# Regular Expression for a single determiner
determiner_pattern = r'\b(?:a|an|the)\b'

# Test
test_string_a = "The cat is on a mat. An apple is tasty. A dog barks."

matches_a = re.findall(determiner_pattern, test_string_a, flags=re.IGNORECASE)
print(matches_a)
```

```python
# Regular Expression for an arithmetic expression
arithmetic_expression_pattern = r'\b(?:\d+[\+\*])*[\d]+\b'

# Test
test_string_b = "2*3+8 is equal to 14. 5+7*2 is 19."

matches_b = re.findall(arithmetic_expression_pattern, test_string_b)
print(matches_b)
```

```python
#Question 3
from urllib import request
from bs4 import BeautifulSoup

def get_text_from_url(url):
    try:
        # Fetch the HTML content
        html = request.urlopen(url).read().decode('utf-8')

        # Parse the HTML using BeautifulSoup and remove markup
        soup = BeautifulSoup(html, 'html.parser')
        text_without_markup = soup.get_text()

        return text_without_markup

    except Exception as e:
        print(f"Error fetching or parsing content from the URL: {e}")
        return None

# Example usage:
url = 'https://www.dsu.edu.pk/contact-us/'
result = get_text_from_url(url)

if result:
    print(result)
```

```python
#Question 5
import re
import nltk
from urllib import request
from bs4 import BeautifulSoup
from nltk.stem import PorterStemmer, LancasterStemmer


def find_phone_numbers_and_emails_tokens(tokens):
    # Join tokens into a single string
    text = ' '.join(tokens)

    # Regex patterns for phone numbers and email addresses
    phone_pattern = re.compile(r'\s*#?\s*(\d{3})[-.\s]*(\d{3})[-.\s]*(\d{4})(?:[-.\s]*(\d+))?\b')
    email_pattern = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b')

    # Find matches
    phone_numbers = [f"{match.group(1)}-{match.group(2)}-{match.group(3)}" if match.group(1) else f"{match.group(2)}-{match.group(3)}-{match.group(4)}" for match in phone_pattern.finditer(text)]
    email_addresses = [match for match in email_pattern.findall(text)]

    return phone_numbers, email_addresses

# URL of the webpage
url = 'https://www.dsu.edu.pk/contact-us/'

# Get text from the URL
text_from_url = get_text_from_url(url)

# Check if text is retrieved successfully
if text_from_url:
    # Tokenize the text using NLTK
    tokens = nltk.word_tokenize(text_from_url)

    # Initialize stemmers
    porter_stemmer = PorterStemmer()
    lancaster_stemmer = LancasterStemmer()

    # Apply stemming
    porter_stems = [porter_stemmer.stem(word) for word in tokens]
    lancaster_stems = [lancaster_stemmer.stem(word) for word in tokens]

    print("Original Tokens:")
    print(tokens)

    # Find phone numbers and email addresses in the original text
    phone_numbers, email_addresses = find_phone_numbers_and_emails_tokens(tokens)
    print("Phone Numbers:")
    for num in phone_numbers:
        print(num)

    print("\nEmail Addresses:")
    for email in email_addresses:
        print(email)

    print("\nPorter Stemmer:")
    print(porter_stems)

    # Find phone numbers and email addresses in the Porter-stemmed text
    phone_numbers, email_addresses = find_phone_numbers_and_emails_tokens(porter_stems)
    print("Phone Numbers:")
    for num in phone_numbers:
        print(num)

    print("\nEmail Addresses:")
    for email in email_addresses:
        print(email)

    print("\nLancaster Stemmer:")
    print(lancaster_stems)

    # Find phone numbers and email addresses in the Lancaster-stemmed text
    phone_numbers, email_addresses = find_phone_numbers_and_emails_tokens(lancaster_stems)
    print("Phone Numbers:")
    for num in phone_numbers:
        print(num)

    print("\nEmail Addresses:")
    for email in email_addresses:
        print(email)
```

**Observations:**<br>
Observing the differences between the two stemmers, you may notice that the Porter Stemmer tends to produce longer stems compared to the Lancaster Stemmer. The Lancaster Stemmer is often more aggressive in its stemming process, leading to potentially shorter stems but at the cost of being more aggressive in its reduction.
